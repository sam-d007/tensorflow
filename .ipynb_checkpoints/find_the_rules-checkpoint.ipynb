### Find the rules
In this notebook, we will create a simple neural network that will find the relationship between two numbers x and y.
Eq : y = 10x - 4
We will provide six values each of x and y for training and then evaluate the model.


```python
import tensorflow as tf
import numpy as np
from tensorflow import keras
```

```python
# model = tf.keras.Sequential()
model = tf.keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
```

```python
# model.add(keras.layers.Dense(units=1, input_shape=[1], dtype=float))
```

```python
#training data
x = np.array([2.0,5.0,-1.0,-3.0,9.0,0.0], dtype=float)
y = np.array([16.0,46.0,-14.0,-34.0,86.0,-4.0], dtype=float)
```

```python
model.compile(loss='mean_squared_error', optimizer='sgd')
```

```python
model.fit(x=x, y=y, epochs=500)
```

```python
model.predict([3]) #actual = 26
```

```python
model.predict([-4]) #actual = -44
```
